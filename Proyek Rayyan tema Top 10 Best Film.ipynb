```python
# 1. Mengurutkan data berdasarkan 'revenue' dari yang terbesar ke terkecil
df_terbesar = df.sort_values(by='revenue', ascending=False)

# 2. Menampilkan 10 film teratas hanya untuk kolom title, year, dan revenue
df_terbesar[['title', 'year', 'revenue']].head(10)
```

```python
# 1. Membersihkan data: Membuang baris yang judul filmnya sama (duplikat)
df_bersih = df.drop_duplicates(subset=['title'])

# 2. Mengurutkan ulang data yang sudah bersih dari revenue terbesar
df_top10 = df_bersih.sort_values(by='revenue', ascending=False)

# 3. Menampilkan 10 film teratas yang sesungguhnya
df_top10[['title', 'year', 'revenue']].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Mengambil hanya 10 film teratas agar grafik tidak berantakan
df_grafik = df_top10.head(10)

# 2. Mengatur gaya visualisasi agar terlihat modern dan bersih
sns.set_theme(style="whitegrid")

# 3. Membuat kanvas grafik (ukuran 10x6 inci)
plt.figure(figsize=(10, 6))

# 4. Membuat Grafik Batang Horizontal (x=revenue, y=title)
# Kita pakai palet warna 'viridis' agar estetik
ax = sns.barplot(
    data=df_grafik,
    x="revenue",
    y="title",
    palette="viridis",
    hue="title", # Untuk mapping warna unik per judul
    legend=False # Mematikan legenda karena judul sudah ada di sumbu Y
)

# 5. Menambahkan Judul dan Label Sumbu dalam Bahasa Indonesia
ax.set_title("Top 10 Film Berdasarkan Pendapatan", fontsize=16, fontweight='bold')
ax.set_xlabel("Pendapatan (Revenue)", fontsize=12)
ax.set_ylabel("Judul Film", fontsize=12)

# 6. Trik Ajaib: Mengubah format angka ribuan (misal 29,000,000 jadi '29M')
import matplotlib.ticker as ticker
ax.xaxis.set_major_formatter(ticker.EngFormatter(unit=''))

# 7. Mengatur tata letak agar rapi dan menampilkan grafik
plt.tight_layout()
plt.show()
```
